## Ranking Search Results by Relevance

## Part 1: Simple Scoring

Score = number of times query appears in text

```python
def calculate_score(query, text):
    """Count how many times query appears."""
    return text.lower().count(query.lower())

# Test
print(calculate_score('fever', 'High fever and low fever'))  # 2
print(calculate_score('fever', 'Patient has rash'))  # 0
```

## Part 2: Score Multiple Documents

```python
documents = [
    'Rubella causes fever',
    'Dengue causes high fever and fever symptoms',
    'Cholera causes diarrhea'
]

query = 'fever'

# Score each document
for doc in documents:
    score = calculate_score(query, doc)
    print(f'Score {score}: {doc}')
```

## Part 3: Sort by Score

```python
# Create list of (score, document) tuples
scored_docs = []
for doc in documents:
    score = calculate_score(query, doc)
    scored_docs.append((score, doc))

print('Before sorting:', scored_docs)

# Sort by score (descending)
scored_docs.sort(reverse=True)

print('After sorting:', scored_docs)
```

## Part 4: Top-K Results

```python
def search_and_rank(query, documents, top_k=3):
    """Search, score, and return top K results."""
    # Score all documents
    scored = []
    for doc in documents:
        score = calculate_score(query, doc)
        if score > 0:  # Only include matches
            scored.append({'score': score, 'text': doc})
    
    # Sort by score descending
    scored.sort(key=lambda x: x['score'], reverse=True)
    
    # Return top K
    return scored[:top_k]

# Test
results = search_and_rank('fever', documents, top_k=2)
for r in results:
    print(f"Score {r['score']}: {r['text']}")
```

## Summary

| Concept | Code |
|---------|------|
| Score | `text.count(query)` |
| Sort | `list.sort(reverse=True)` |
| Top-K | `list[:k]` |
